# Comparing auxilary metrics

```python
import prepare_data
url2record, groups, url2group, markups = prepare_data.prepare_data(train=True)
```

```python
from evaluate_metric import evaluate_metric

def matrix_func(matrix):
    return matrix.T
```

Evaluating rouge as auxillary metric:

```python
from rouge import Rouge
rouge_metric = Rouge()

print(evaluate_metric(groups, markups, url2record, rouge_metric, 'pagerank', matrix_func=matrix_func))
```

Evaluating bertscore with 'ai-forever/sbert_large_mt_nlu_ru' model as auxillary metric:

```python
from bertscore import Bertscore

bertscore_metric = Bertscore('ai-forever/sbert_large_mt_nlu_ru', num_layers=16)
bertscore_metric.set_metric_type('f1')

print(evaluate_metric(groups, markups, url2record, bertscore_metric, 'pagerank', matrix_func=matrix_func))
```

# Eval

```python
import prepare_data
eval_groups, eval_markups, url2record, best_headlines, nont_clusters = prepare_data.prepare_comp_data(min_size=5)
```

```python
from rouge import Rouge
rouge_metric = Rouge()
```

```python
# rouge
from compare import compare_metric, pagerank_scorer
print(compare_metric(eval_groups, eval_markups, url2record, rouge_metric, best_headlines,
                     nont_clusters))
```

```python
pagerank_rouge = pagerank_scorer(rouge_metric)
print(compare_metric(eval_groups, eval_markups, url2record, pagerank_rouge, best_headlines,
                     nont_clusters))
```

```python
print(compare_metric(eval_groups, eval_markups, url2record, bertscore_metric, best_headlines,
                     nont_clusters))
```

```python
pagerank_bertscore = pagerank_scorer(bertscore_metric)
print(compare_metric(eval_groups, eval_markups, url2record, pagerank_bertscore, best_headlines,
                     nont_clusters))
```
